```python
import pandas as pd
import requests
import time 
import numpy as np
```

```python
species_df = pd.read_csv('species_cleaned_wiki2.csv', na_values=['', ' ', 'NA', 'NaN', 'none', 'null'])
species_df
```

# Adding additional information on taxa from GBIF API. Using python wrapper pygbif.

```python
from pygbif import species
```

res = species.name_backbone(name='Antigone antigone')
print(res)

```python
kingdoms = []
phyla = []
classes = []
orders = []
families = []
genera = []
species_list = []
usageKeys = [] 
scientific_names = []
canonical_names = []
ranks = []
taxonomic_status = []

for taxon in species_df['taxon']:
    try:
        result = species.name_backbone(name=taxon)
        kingdoms.append(result.get('kingdom'))
        phyla.append(result.get('phylum'))
        classes.append(result.get('class'))
        orders.append(result.get('order'))
        families.append(result.get('family'))
        genera.append(result.get('genus'))
        species_list.append(result.get('species'))
        usageKeys.append(result.get('usageKey'))
        scientific_names.append(result.get('scientificName'))
        canonical_names.append(result.get('canonicalName'))
        ranks.append(result.get('rank'))
        taxonomic_status.append(result.get('status'))

        print(f"Found: {result.get('scientificName')} | usageKey: {result.get('usageKey')}")
        time.sleep(1)
        
    except Exception as e:
        print(f"Error fetching {taxon}: {e}")
        time.sleep(1)
        kingdoms.append(None)
        phyla.append(None)
        classes.append(None)
        orders.append(None)
        families.append(None)
        genera.append(None)
        species_list.append(None)
        usageKeys.append(None)
        scientific_names.append(None)
        canonical_names.append(None)
        ranks.append(None)
        taxonomic_status.append(None)

        
            

species_df['kingdom'] = kingdoms
species_df['phylum'] = phyla
species_df['class'] = classes
species_df['order'] = orders
species_df['family'] = families
species_df['genus'] = genera 
species_df['species_gbif'] = species_list
species_df['usageKey'] = usageKeys
species_df['scientificName'] = scientific_names
species_df['canonicalName'] = canonical_names
species_df['rank'] = ranks
species_df['status'] = taxonomic_status
```

```python
species_df
```

```python
#Saving results to CSV 
species_df.to_csv("species_with_gbif.csv", index=False)
```

```python
#Trying to get additional info from GBIF
extincts = []
threat_statuses = []
authorships = []
habitats_list = []
synonyms = []

for key in species_df['usageKey']:
    if pd.isna(key):
        extincts.append(None)
        threat_statuses.append(None)
        authorships.append(None)
        habitats_list.append(None)
        synonyms.append(None)
        continue

    key = int(key)

    # Main species info
    try:
        response = requests.get(f"https://api.gbif.org/v1/species/{key}")
        data = response.json()

        extincts.append(data.get("extinct"))
        threat_statuses.append(data.get("threatStatus"))
        authorships.append(data.get("authorship"))
        habitats_list.append(data.get("habitats"))
        synonyms.append(data.get("synonym"))

        print(f"Found: {data.get('scientificName')}")

    except Exception as e:
        extincts.append(None)
        threat_statuses.append(None)
        authorships.append(None)
        habitats_list.append(None)
        synonyms.append(None)
        print(f"Error for usageKey {key}: {e}")

    time.sleep(0.5)  

# Add enriched fields to the DataFrame
species_df["extinct"] = extincts
species_df["threatStatus"] = threat_statuses
species_df["authorship"] = authorships
species_df["habitats"] = habitats_list
species_df["synonym"] = synonyms
```

```python
#GBIF did not have data on threatStatus, habitats etc. Dropping these columns
species_df = species_df.drop(columns = ['threatStatus', 'habitats', 'synonym'])
```

```python
species_cleaned = species_df.drop(columns = ['usageKey'])
```

```python
species_cleaned
```

```python
# Extracting year and author from GBIF
species_cleaned["year_gbif"] = species_cleaned["authorship"].str.extract(r"(\d{4})")
```

```python
species_cleaned["author_gbif"] = (species_cleaned["authorship"].str.extract(r"^(.*?),\s*\d{4}").iloc[:, 0].str.replace(r"[()]", "", regex=True).str.strip())                
```

```python
#Checking status
species_cleaned['status'].value_counts()
```

```python
#Filtering to drop doubtful names.
species_accepted = species_cleaned[species_cleaned['status'] == 'ACCEPTED']
species_synonym = species_cleaned[species_cleaned['status'] == 'SYNONYM']
```

```python
species_synonym
```

```python
#Manually checked the synonyms to see if accepted names still include the fictional inspiration, and creating a list of those to keep.
to_keep = ['Antigone antigone', 'Pseudione quasimodo', 'Albicetus oxymycterus', 'Meoneura tinkerbellae', 'Urostephanus lokii', 'Rathalos treecko', 'Litoria sauroni', 'Urostephanus balrog', 'Meoneura hobbitoides', 'Borophagus orc', 'Pycnophyes smaug', 'Meoneura yodai']
to_keep_df = species_synonym[species_synonym['taxon'].isin(to_keep)]
species_accepted = pd.concat([species_accepted, to_keep_df], ignore_index=True)
```

```python
#These two have been classified as genus, except the species part of the scientific name is derived from fictional works, not the genus. Reclassifying for analysis purpose.
species_accepted.loc[species_accepted['taxon'] == 'Urostephanus lokii', 'rank'] = 'SPECIES'
species_accepted.loc[species_accepted['taxon'] == 'Urostephanus balrog', 'rank'] = 'SPECIES'
```
